```python
# Import the Pandas package
import pandas as pd
import numpy as np
import re 
def clean_text(string,replace='_'):
    regex = '[^A-Za-z0-9]+'
    return re.sub(regex, replace, string)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")  # Suppress all warnings
```

```python
dfPI = pd.read_csv("MSAPersonalIncome/MSAPersonalIncome.csv", skiprows = 4, skipfooter = 11)
dfPI['city']= dfPI['GeoName'].str.split('(',expand = True)[0]
years = dfPI.iloc[:,4:23].columns.values.tolist()
dfPIclean = pd.melt(dfPI, id_vars=['city','Description'],  value_vars=years,
        var_name=('Year'), value_name='Values')
```

```python
def replace_emdash(string):
    return string.replace('â€“','-',100)

def replace_all_special(string):
    string = re.sub('[^A-Za-z, ]+', '', string.lower())
    return re.sub('[^A-Za-z,]+', '_', string.strip())

def split_name(string):
    split = string.split(',')
    if(len(split)>1):
        city= split[0].strip()
        state = split[1].strip()
        city_split=city.split('-')
        state_split=state.split('-')
        return city_split[0]+', '+state_split[0]
    else:
        return split[0].strip()
```

```python
dfPIclean['city'] = dfPIclean['city'].apply(split_name)
dfPIclean['Description'] = dfPIclean['Description'].apply(str.strip).apply(replace_all_special)
dfPIclean
```

```python
dfPIclean = pd.pivot_table(dfPIclean, values='Values', index=['city','Year'],columns=['Description']).reset_index()
dfPIclean.to_csv('CleanData/DPI_Clean.csv',index=False)
```
